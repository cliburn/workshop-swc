**Example 2: Exact M.W. of ssDNA (e.g., Oligonucleotides)**

Suppose we are give a ssDNA sequence 'GATTACA'. We want to write a function calculate the molecular weight of any such sequence using the following formula: 

M.W. = (An x 313.2) + (Tn x 304.2) + (Cn x 289.2) + (Gn x 329.2) + 79.0ª
An, Tn, Cn, and Gn are the number of each respective nucleotide within the polynucleotide.
ªAddition of "79.0" to the M.W. takes into account the 5' monophosphate left by most restriction enzymes. No phosphate is present at the 5' end of strands made by primer extension.

<http://www.lifetechnologies.com/us/en/home/references/ambion-tech-support/rna-tools-and-calculators/dna-and-rna-molecular-weights-and-conversions.html>

```python
dna_mw = {
    'A' : 313.2,
    'T' : 304.2, 
    'C' : 289.2,
    'G' : 329.2
    }

def mw(seq, mw_dict, offset=79.04):
    """Return molecular weight of the ssDNA string seq."""
    wt = sum(mw_dict[nuc] for nuc in seq) + offset
    return wt

dna = 'GATTACA'
print mw(dna, dna_mw)
```

**Example 3: Find the frequncy of a specific DNA sequence in a collection of many DNA sequences (e.g. from fragmentation)**

Suppose we have sequences from 1 million short DNA fragments. We want to answer some of the following questions:

1. What is the frequency of the fragment 'GATTACA'?
2. What is the distribution of fragment lengths?
3. What are the 3 most common fragments?

```python

# Question 1 - manual solution
def count_fragments(frags):
    """Return a dictionary where the key is the fragment sequence and values is number of times it is found."""
    counts = {}
    for frag in frags:
        counts[frag] = counts.get(frag, 0) + 1
    return counts

counts = count_fragments(frags)
print counts['GATTACA']

# Question 2 - using collections.Counter
frag_lens = [len(frag) for frag in frags]
from collections import Counter
print Counter(frag_lens)

# Question 3 - using collections.Counter
from collections import Counter
print Counter(frags).most_common(3)
```

### Example 4: Finding ORFs and transcription products

Suppose we are given a DNA sequence. 

1. Can we find an open reading frame within the sequence?
2. What would the amino acid sequence be of the protein translated from this ORF?

```python
# Question 1
def find_orf(seq):
    """Returns list of codons in ORF or the empty list if an ORF is not found."""
      
    start_codon = 'AUG'
    stop_codons = ['TAG', 'TAA', 'TGA']

    stop_idx = 0
    start_idx = seq.find(start_codon)
    
    # if start codon is not found, there is no ORF so return empty list
    if start_idx == -1:
        return []
    
    # possile codons are the 3-mers beginning with the start codon
    possible_codons = [seq[i:(i+3)] for i in range(start_idx, len(seq), 3)]

    # find stop codon if it is among possibe codons and return (possible empty) list of codons
    stop = 0
    for stop_codon in stop_codons:
        if stop_codon in possible_codons:
            stop = possible_codons.index(stop_codon)

    return possible_codons[:stop]

# DNA string is left flank + start + bunch of codons + stop + right flank
dna = 'GT' + 'AUG' + 'CCACGTTCCCATATCAGTGAT' + 'TGA' + 'TA'
orf = find_orf(dna)

# Question 2
from seq_utils import codon_map

protein = [codon_map[codon] for codon in orf]
print protein
```
