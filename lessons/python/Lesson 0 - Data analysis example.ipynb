### Data analysis example

We read in 13 comma-separated value (CSV) files each containing data on 60 fictitious patients. In each file, one row correponds to tempeature measurements over 40 days for one patient, with one temperature measurement per day. For each file, we want to see a heatmap of the temperature distribution over 40 days for each of the 60 patients, and also to plot the average temperature across all patients on each day.

```python
def display(data, nrows, i):
    """Plot one row of data - heatmap + line plot."""
    subplot(nrows, 2, 2*i+1)
    imshow(data)
    subplot(nrows, 2, 2*i+2)
    plot(data.mean(0))
 
# read in daata sets
datasets = [pd.read_csv(f, header=None) for f in glob.glob('inf*csv')]
display_all(datasets)

# use a loop to display all dataset
nrows = len(datasets)
panel_width = 3
panel_height = 2
figure(figsize=(2*panel_width, nrows*panel_height))
for i, dataset in enumerate(datasets):
    display(dataset, nrows, i)
tight_layout()
```

```python
# We can read in CSV files with the pandas module, and specify that there is no header row
df = pd.read_csv('inflammation-01.csv', header=None)
```

```python
# check the number of rows (=60 patinets) and cols (=40 measurements per patient)
df.shape
```

```python
# and then we use "slice" notation to show the first 5 rows
# slice notnation is [start:stop:step]]
df[:5]
```

```python
# To calculate the average cross rows, we use the mean() method of the dataframe.
# The parameter axis=0 means to do the averaging across all 60 rows (patients), 
# giving one average value per column (day)
# If we set axis=1, we will get 60 values, one for each patient, 
# where each value is the average temperature for that paient over all 40 days.
df.mean(axis=0)
```

```python
# The heatmap for an array of numbers can be plotted using imshow
imshow(df);
```

```python
# Similarly the average temperature per day can be plotted with plot
plot(df.mean(0));
```

```python
# To combine plots, we use the subplot function with arguments subplot(nrows, ncols, index)
subplot(1,2,1)
imshow(df)
subplot(1,2,2)
plot(df.mean(0));
```

```python
# Since we want to repeat drawing this combinaiton of panels many times,
# we put it into a function

def display(data, nrows, i):
    """Plot one row of data - heatmap + line plot."""
    subplot(nrows, 2, 2*i+1)
    imshow(data)
    subplot(nrows, 2, 2*i+2)
    plot(data.mean(0))
```

```python
# To read in all 13 files, we use the glob function from the glob module
glob.glob('inf*csv')
```

```python
# We use a list comprehension to load each file into a dataframe
# list comprehsnsons are "short forms" for looping to generate a colleciton of items
datasets = [pd.read_csv(f) for f in glob.glob('inf*csv')]
```

```python
# Finally, we use a for loop to iterate over datasets
# and call the display function for each dataset
# Plus some graphics directives to specify the plot layout etc (not shown)

for i, dataset in enumerate(datasets):
    display(dataset, nrows, i)
```
